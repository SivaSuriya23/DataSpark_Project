```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser

# Convert Open Date column to datetime using dateutil parser
df['Open Date'] = df['Open Date'].apply(parser.parse)

# Distribution of Square Meters
plt.figure(figsize=(10, 6))
sns.histplot(df['Square Meters'], bins=30, kde=True)
plt.title('Distribution of Square Meters')
plt.xlabel('Square Meters')
plt.ylabel('Frequency')
plt.show()

# Store Open Dates Over Time
plt.figure(figsize=(12, 8))
sns.lineplot(x='Open Date', y='Square Meters', hue='Country', data=df)
plt.title('Store Open Dates Over Time')
plt.xlabel('Open Date')
plt.ylabel('Square Meters')
plt.legend(title='Country')
plt.show()

# Correlation Heatmap
corr_df = df[['StoreKey', 'Square Meters']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_df, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot of Numerical Features
sns.pairplot(df[['StoreKey', 'Square Meters']])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()

# Distribution of Stores by Country
plt.figure(figsize=(10, 6))
sns.countplot(x='Country', data=df)
plt.title('Distribution of Stores by Country')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.show()

# Average Square Meters by Country
avg_square_meters_by_country = df.groupby('Country')['Square Meters'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Square Meters', data=avg_square_meters_by_country)
plt.title('Average Square Meters by Country')
plt.xlabel('Country')
plt.ylabel('Average Square Meters')
plt.show()
```
