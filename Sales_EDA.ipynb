```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser

# Convert Date columns to datetime using dateutil parser
df['Order Date'] = df['Order Date'].apply(parser.parse)
df['Delivery Date'] = df['Delivery Date'].apply(lambda x: parser.parse(x) if pd.notnull(x) else x)

# Distribution of Quantity
plt.figure(figsize=(10, 6))
sns.histplot(df['Quantity'], bins=30, kde=True)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# Orders Over Time
plt.figure(figsize=(12, 8))
sns.lineplot(x='Order Date', y='Quantity', hue='Currency Code', data=df)
plt.title('Orders Over Time')
plt.xlabel('Order Date')
plt.ylabel('Quantity')
plt.legend(title='Currency Code')
plt.show()

# Delivery Dates Analysis
plt.figure(figsize=(12, 8))
sns.histplot(df['Delivery Date'].dropna(), bins=30, kde=True)
plt.title('Distribution of Delivery Dates')
plt.xlabel('Delivery Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Correlation Heatmap
corr_df = df[['Order Number', 'CustomerKey', 'StoreKey', 'ProductKey', 'Quantity']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_df, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot of Numerical Features
sns.pairplot(df[['Order Number', 'CustomerKey', 'StoreKey', 'ProductKey', 'Quantity']])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()

# Distribution of Orders by Currency Code
plt.figure(figsize=(10, 6))
sns.countplot(x='Currency Code', data=df)
plt.title('Distribution of Orders by Currency Code')
plt.xlabel('Currency Code')
plt.ylabel('Frequency')
plt.show()

# Average Quantity by Currency Code
avg_quantity_by_currency = df.groupby('Currency Code')['Quantity'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Currency Code', y='Quantity', data=avg_quantity_by_currency)
plt.title('Average Quantity by Currency Code')
plt.xlabel('Currency Code')
plt.ylabel('Average Quantity')
plt.show()
```
