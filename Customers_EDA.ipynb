```python
# Check the column names
print(df.columns)

# Rename columns if necessary
df.rename(columns=lambda x: x.strip(), inplace=True)

# Change 'CustomerKey' to int
df['CustomerKey'] = df['CustomerKey'].astype(int)

# Change 'Gender' to category
df['Gender'] = df['Gender'].astype('category')

# Change 'Name', 'City', 'State', 'Country', 'Continent' to string (object in pandas)
df['Name'] = df['Name'].astype(str)
df['City'] = df['City'].astype(str)
df['State'] = df['State'].astype(str)
df['Country'] = df['Country'].astype(str)
df['Continent'] = df['Continent'].astype(str)

# Change 'State Code' to category
df['State Code'] = df['State Code'].astype('category')

# Clean and convert 'Zip Code' to string
df['Zip Code'] = df['Zip Code'].apply(lambda x: ''.join(filter(str.isdigit, str(x))))

# Convert 'Birthday' to datetime
df['Birthday'] = pd.to_datetime(df['Birthday'], errors='coerce')

# View the updated data types
print(df.dtypes)
```

```python
# Function to parse dates and retain original value if parsing fails
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, errors='raise')
    except Exception:
        return date_str

# Apply the function to the 'Birthday' column
df['Birthday'] = df['Birthday'].apply(parse_date)

# Check the data types
print(df.dtypes)
```

```python
# Handling Outliers

# Calculate IQR for the 'CustomerKey' column
Q1 = df['CustomerKey'].quantile(0.25)
Q3 = df['CustomerKey'].quantile(0.75)
IQR = Q3 - Q1

# Determine outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['CustomerKey'] < lower_bound) | (df['CustomerKey'] > upper_bound)]
print("Outliers in 'CustomerKey':")
print(outliers)
```

```python
# Univariate Analysis
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of numerical variable 'CustomerKey'
sns.histplot(df['CustomerKey'], kde=True)
plt.show()

# Distribution of categorical variable 'Gender'
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
# Bivariate Analysis
# Numerical vs. Numerical: Scatter plot of 'CustomerKey' and 'Zip Code' (converted to numerical)
df['Zip Code Num'] = pd.to_numeric(df['Zip Code'], errors='coerce')
sns.scatterplot(x='CustomerKey', y='Zip Code Num', data=df)
plt.show()

# Numerical vs. Categorical: Box plot of 'CustomerKey' across 'Gender'
sns.boxplot(x='Gender', y='CustomerKey', data=df)
plt.show()
```

```python
# Hypothesis Testing
from scipy.stats import ttest_ind

# Hypothesis Testing: Compare 'Age' between 'Male' and 'Female'
group1 = df[df['Gender'] == 'Male']['Age']
group2 = df[df['Gender'] == 'Female']['Age']
t_stat, p_value = ttest_ind(group1, group2, nan_policy='omit')
print(f'T-statistic: {t_stat}, P-value: {p_value}')
```
