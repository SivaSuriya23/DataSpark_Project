```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Unit Cost and Unit Price to numeric values
df['Unit Cost USD'] = df['Unit Cost USD'].replace('[\$,]', '', regex=True).astype(float)
df['Unit Price USD'] = df['Unit Price USD'].replace('[\$,]', '', regex=True).astype(float)

# Distribution of Unit Costs
plt.figure(figsize=(10, 6))
sns.histplot(df['Unit Cost USD'], bins=10, kde=True)
plt.title('Distribution of Unit Costs')
plt.xlabel('Unit Cost USD')
plt.ylabel('Frequency')
plt.show()

# Distribution of Unit Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Unit Price USD'], bins=10, kde=True)
plt.title('Distribution of Unit Prices')
plt.xlabel('Unit Price USD')
plt.ylabel('Frequency')
plt.show()

# Unit Cost vs. Unit Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Unit Cost USD', y='Unit Price USD', hue='Color', data=df)
plt.title('Unit Cost vs. Unit Price')
plt.xlabel('Unit Cost USD')
plt.ylabel('Unit Price USD')
plt.legend(title='Color')
plt.show()

# Correlation Heatmap
corr_df = df[['ProductKey', 'Unit Cost USD', 'Unit Price USD', 'SubcategoryKey', 'CategoryKey']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_df, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot of Numerical Features
sns.pairplot(df[['Unit Cost USD', 'Unit Price USD']])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()

# Distribution of Products by Color
plt.figure(figsize=(10, 6))
sns.countplot(x='Color', data=df)
plt.title('Distribution of Products by Color')
plt.xlabel('Color')
plt.ylabel('Frequency')
plt.show()

# Average Unit Cost by Color
avg_cost_by_color = df.groupby('Color')['Unit Cost USD'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Color', y='Unit Cost USD', data=avg_cost_by_color)
plt.title('Average Unit Cost by Color')
plt.xlabel('Color')
plt.ylabel('Average Unit Cost USD')
plt.show()

# Average Unit Price by Color
avg_price_by_color = df.groupby('Color')['Unit Price USD'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Color', y='Unit Price USD', data=avg_price_by_color)
plt.title('Average Unit Price by Color')
plt.xlabel('Color')
plt.ylabel('Average Unit Price USD')
plt.show()
```
