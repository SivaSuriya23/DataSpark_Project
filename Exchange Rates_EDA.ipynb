```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser

# Convert Date column to datetime using dateutil parser
df['Date'] = df['Date'].apply(parser.parse)

# Distribution of Exchange Rates
plt.figure(figsize=(10, 6))
sns.histplot(df['Exchange'], bins=30, kde=True)
plt.title('Distribution of Exchange Rates')
plt.xlabel('Exchange Rate')
plt.ylabel('Frequency')
plt.show()

# Exchange Rates Over Time
plt.figure(figsize=(12, 8))
sns.lineplot(x='Date', y='Exchange', hue='Currency', data=df)
plt.title('Exchange Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend(title='Currency')
plt.show()

# Correlation Heatmap
pivot_df = df.pivot(index='Date', columns='Currency', values='Exchange')
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Exchange Rates')
plt.show()

# Pairplot of Exchange Rates
sns.pairplot(pivot_df)
plt.suptitle('Pairplot of Exchange Rates', y=1.02)
plt.show()
```
